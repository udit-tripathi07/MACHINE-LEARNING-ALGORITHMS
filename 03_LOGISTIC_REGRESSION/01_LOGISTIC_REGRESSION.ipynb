```python
import matplotlib.pyplot as plt  
import numpy as np
```

```python
np.random.seed(0)

X1 = np.random.randn(50,2) + [-3,0]
X2 = np.random.randn(50,2) + [3,0]

X = np.vstack((X1,X2))
y = np.array([0]*50 + [1]*50)
```

```python
plt.scatter(X[:,0],X[:,1],c=y,s=100)
```

```python
def step(z):
    if(z>0): return 1
    return 0
```

```python
def perceptron(X,y):

    X=np.insert(X,0,1,axis=1)
    weights=np.ones(X.shape[1])   # take the coefficients of a random line
    lr=0.1

    for i in range(1000):
        j=np.random.randint(0,100)   # choose random number then check for that point
        y_hat=step(np.dot(X[j],weights))
        weights=weights+lr*(y[j]-y_hat)*X[j]

    return weights[0],weights[1:]
```

```python
intercept_,coef_=perceptron(X,y)
```

```python
print(coef_)
print(intercept_)
```

```python
m=-(coef_[0]/coef_[1])
b=-(intercept_/coef_[1])
```

```python
x_in=np.linspace(-3,3,50)
y_in=m*x_in+b      # equation of our line 
```

```python
plt.plot(x_in,y_in,color='red')
plt.scatter(X[:,0],X[:,1],c=y)
plt.ylim(-3,3)
```

```python
plt.plot(x_in,y_in,color='red')
plt.plot(x_in1,y_in1,color='black')
plt.scatter(X[:,0],X[:,1],c=y)
plt.ylim(-3,2)
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def perceptron(X,y):

    X=np.insert(X,0,1,axis=1)
    weights=np.ones(X.shape[1])   # take the coefficients of a random line
    lr=0.1

    for i in range(1000):
        j=np.random.randint(0,100)   # choose random number then check for that point
        y_hat=sigmoid(np.dot(X[j],weights))
        weights=weights+lr*(y[j]-y_hat)*X[j]

    return weights[0],weights[1:]
```
