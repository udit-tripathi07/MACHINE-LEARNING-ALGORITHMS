***MULTINOMIAL LOGISTIC REGRESSION USING SOFTMAX***

```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
encoder=LabelEncoder()
df['species']=encoder.fit_transform(df['species'])
df
```

```python
df=df[['sepal_length','petal_length','species']] # we work on only two column inorder to visualize graph
df.head()
```

```python
X=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
lor=LogisticRegression(solver='lbfgs')
```

```python
lor.fit(X_train,y_train)
```

```python
y_pred=lor.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print(accuracy_score(y_test,y_pred) )
```

```python
lor.predict_proba([[5,3]])
# gives the probability of that value 
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X.values, y.values, lor, legend=2)
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.title('Softmax on Iris')
```
