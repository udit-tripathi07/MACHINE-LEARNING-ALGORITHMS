```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_features=5,n_redundant=0,n_informative=5,n_clusters_per_class=1)
```

```python
df=pd.DataFrame(X,columns=['C1','C2','C3','C4','C5'])
df['target']=y
df.head()
```

```python
# for row sampling
def sample_rows(df,percent):
    return df.sample(int(percent*df.shape[0]),replace=True)
```

```python
df1=sample_rows(df,0.1) # RANDOM 10 SAMPLES
df2=sample_rows(df,0.1)
df3=sample_rows(df,0.1)
```

```python
df1
```

```python

from sklearn.tree import DecisionTreeClassifier
clf1 = DecisionTreeClassifier()
clf2 = DecisionTreeClassifier()
clf3 = DecisionTreeClassifier()
```

```python

clf1.fit(df1.iloc[:,0:5],df1.iloc[:,-1])
clf2.fit(df2.iloc[:,0:5],df2.iloc[:,-1])
clf3.fit(df3.iloc[:,0:5],df3.iloc[:,-1])
```

```python
clf1.predict(np.array([2.339910	,-1.565003	,-0.815188	,0.412943	,0.522091]).reshape(1,5))
```

```python
clf2.predict(np.array([2.339910	,-1.565003	,-0.815188	,0.412943	,0.522091]).reshape(1,5))
```

```python
clf3.predict(np.array([2.339910	,-1.565003	,-0.815188	,0.412943	,0.522091]).reshape(1,5))
```
