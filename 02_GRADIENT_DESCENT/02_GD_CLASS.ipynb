**WE FIND THE VALUE OF B USING GRADIENT DESCENT CLASSS WHERE M IS KNOWN**

```python
import numpy as np
from sklearn.datasets import make_regression
```

```python
X,y=make_regression(n_samples=100,n_features=1,n_informative=1,n_targets=1,noise=20)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(4,3))
plt.scatter(X,y)
```

**USING OLS**

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X,y)
print(lr.coef_)
print(lr.intercept_)
```

```python
plt.figure(figsize=(4,3))
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='crimson')
```

**OUR OWN CLASS**

```python
m=26.57
```

```python
class GDRe:
 def __init__(self,learning_rate,epochs):
  self.m=26.57
  self.b=-100 # random  
  self.ler=learning_rate
  self.epochs=epochs

 def fit(self,X,y):
   for i in range(self.epochs):
    loss_slope=-2*np.sum(y-self.m*X.ravel()-self.b)
    self.b=self.b-(self.ler*loss_slope)
    print(loss_slope,self.b)
   print(self.b)
```

```python
gd=GDRe(0.1,10)
```

```python
gd.fit(X,y)
# you can see our b is very very high because of the higher learning rate
```

```python
gd=GDRe(0.01,10)
gd.fit(X,y)
```

```python
gd=GDRe(0.001,50)
gd.fit(X,y)
```
