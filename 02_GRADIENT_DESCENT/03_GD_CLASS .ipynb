**WE FIND THE VALUE OF BOTH M AND B USING GRADIENT DESCENT**

```python
import numpy as np
from sklearn.datasets import make_regression
```

```python
X,y=make_regression(n_samples=100,n_features=1,n_informative=1,n_targets=1,noise=20,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(4,3))
plt.scatter(X,y)
```

**USING OLS**

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X,y)
print(lr.coef_)
print(lr.intercept_)
```

```python
plt.figure(figsize=(4,3))
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='crimson')
```

**OUR OWN CLASS**

```python
class GDRe:
 def __init__(self,learning_rate,epochs):
  self.m=80  # random
  self.b=-100 # random  
  self.ler=learning_rate
  self.epochs=epochs

 def fit(self,X,y):
   for i in range(self.epochs):
     loss_slope_b=-2*np.sum(y-self.m*X.ravel()-self.b)
     self.b=self.b-(self.ler*loss_slope_b)

     loss_slope_m=-2*np.sum((y-self.m*X.ravel()-self.b)*X.ravel())
     self.m=self.m-(self.ler*loss_slope_m)
    
   print(self.m,self.b)
```

```python
gd=GDRe(0.01,100)
```

```python
gd.fit(X,y)
# m and b nearly same as OLS 
```
