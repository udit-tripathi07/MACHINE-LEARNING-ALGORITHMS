***GEOMETRICAL INTUITION OF SIMPLE LINEAR REGRESSION***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
pc=pd.read_csv('placement.csv')
pc.head()
```

```python
plt.figure(figsize=(5,4))
plt.scatter(pc['cgpa'],pc['package'],color='crimson')
plt.xlabel("CGPA",color='blue')
plt.ylabel("PACKAGE",color='blue')
```

```python
X=pc.iloc[:,0:1]  # cgpa
y=pc.iloc[:,-1]   # package
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
X_test.head()
```

```python
y_test.head()
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
# actual value is 4.10 but our model gives 3.89
```

```python
plt.figure(figsize=(5,4))
plt.scatter(pc['cgpa'],pc['package'],color='blue')
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel("CGPA")
plt.ylabel("PACKAGE")
```

```python
# we find the slope and y-intercept of the above line
m=lr.coef_ 
m   
```
