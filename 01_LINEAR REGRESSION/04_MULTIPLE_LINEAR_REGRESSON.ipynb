**BUILDING OUR OWN MODEL CLASS**

```python
class Udit_LR:
 def __init__(self):
  self.coef=None
  self.intercept_=None
 def fit(self,X_train,y_train):
  X_train= np.insert(X_train,0,1,axis=1)  # adding 1 to each row of our data for matrix

  beta=np.linalg.inv(np.dot(X_train.T,X_train)).dot(X_train.T).dot(y_train) #beta=inv(XT*X)*XT*Y
  self.intercept_=beta[0]
  self.coef_=beta[1:]
  print("INTERCEPT:")
  print(self.intercept_)
  print("COEF:")
  print(self.coef_)
  
 def predict(self,X_test):
  y_pred= self.intercept_+np.dot(X_test,self.coef_)
  return y_pred
```

```python
lr=Udit_LR()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.predict(X_test)
```
