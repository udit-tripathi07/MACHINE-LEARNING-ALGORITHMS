```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Mall_Customers.csv')
```

```python
df.head()
```

```python
X=df.iloc[:,3].values
y=df.iloc[:,4].values
```

```python
plt.scatter(X,y)
```

```python
X=df.iloc[:,3:].values
```

```python
from sklearn.cluster import KMeans
```

```python
# now we fing how many clusters we have to make
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit(X)
    wcss.append(km.inertia_) # distance from centroid of each cluster to every point
```

```python
plt.plot(range(1,11),wcss)
```

```python
# let's take clusters =6
km=KMeans(n_clusters=5)
```

```python
y_means=km.fit_predict(X)
```

```python
y_means
```
